```python
products=soup.find_all('div',class_='_4ddWXP')
products
```

```python
import pandas as pd

Product_name = []
Ratings = []
No_of_reviews = []

# Iterate through the products and extract data
for product in products:
    product_name_elem = product.find("a", class_="s1Q9rs")
    ratings_elem = product.find('div', class_='_3LWZlK')
    reviews_elem = product.find('span', class_='_2_R_DZ')
    
    if product_name_elem:
        Product_name.append(product_name_elem.text.strip())
    else:
        Product_name.append(None)  # Handle the case where the element is not found
    
    if ratings_elem:
        Ratings.append(ratings_elem.text.strip())
    else:
        Ratings.append(None)
    
    if reviews_elem:
        No_of_reviews.append(reviews_elem.text.strip('()'))
    else:
        No_of_reviews.append(None)

# Create a DataFrame
df = pd.DataFrame({
    'Product Name': Product_name,
    'Ratings': Ratings,
    'No of Reviews': No_of_reviews
})
```

# all pages

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

Product_name = []
Ratings = []
No_of_reviews = []

for j in range(1, 300):
    search_string = 'eco product'.replace(" ", "")
    flipkart_url = 'https://www.flipkart.com/search?q=' + search_string + f'&page={j}'

    headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.0.0 Safari/537.36'
     }

    url = requests.get(flipkart_url, headers=headers)
    soup = BeautifulSoup(url.text, 'html.parser')
    products = soup.find_all('div', class_='_4ddWXP')

    for product in products:
        product_name_elem = product.find("a", class_="s1Q9rs")
        ratings_elem = product.find('div', class_='_3LWZlK')
        reviews_elem = product.find('span', class_='_2_R_DZ')

        if product_name_elem:
            Product_name.append(product_name_elem.text.strip())
        else:
            Product_name.append(None)

        if ratings_elem:
            Ratings.append(float(ratings_elem.text.strip()))
        else:
            Ratings.append(None)

        if reviews_elem:
            no_of_reviews_text = reviews_elem.text.strip('()').replace(',', '')  # Remove the comma
            No_of_reviews.append(int(no_of_reviews_text))
        else:
            No_of_reviews.append(None)

# Create a DataFrame
df = pd.DataFrame({
    'Product Name': Product_name,
    'Ratings': Ratings,
    'No of Reviews': No_of_reviews
})
```
